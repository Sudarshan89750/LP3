```python
x=2
lr=0.01
step=1
precision=0.000001
max_iter=10000
iters=0
gf=lambda x:(x+3)**2
```

```python
gd=[]
```

```python
while precision<step and iters<max_iter:
    prev=x
    x-=lr*gf(prev)
    step=abs(x-prev)
    iters+=1
    print("iteration: ",iters,' values :',x)
    gd.append(x)
print('local Minima',x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
